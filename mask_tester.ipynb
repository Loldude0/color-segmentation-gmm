```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

train = []

# Load images and convert to HSL color space
for image in os.listdir('./test_images'):
    img = cv2.imread('./test_images/' + image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)  # Convert to HSL
    if img.shape[0] > 480:
        #rotate
        img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    print(img.shape)
    train.append(img)

train = np.array(train)

orange = []

# Loop through each image
for img in train:

    lower = np.array([0, 00, 120]) 
    upper = np.array([30, 240, 255]) 
    
    mask = cv2.inRange(img, lower, upper)  # Apply mask
    result = cv2.bitwise_and(img, img, mask=mask)  # Masked result
    orange.append(result)

for i in range(len(orange)):
    plt.figure(figsize=(15,5))
    plt.subplot(131)
    plt.imshow(train[i])
    plt.subplot(132)
    plt.imshow(orange[i])
    plt.show()
```
